# Business Problem 
- Understanding The covid- 19 Dataset and presenting a visualization for easy understanding

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.simplefilter('ignore')
```

```python
df = pd.read_csv('country_wise_latest.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Country/Region'].unique()
```

```python
df['WHO Region'].unique()
```

```python
# Visualizing various Regions's Confirmed cases, Recovered & Deaths
#plt.figure(figsize =(10,10))
plt.bar(df['WHO Region'],df['Confirmed'],label='Confirmed Cases')
plt.bar(df['WHO Region'],df['Recovered'],label='Recovered')
plt.bar(df['WHO Region'],df['Deaths'],label='Died')
plt.title('Covid Cases in countries')
plt.xlabel('Region')
plt.ylabel('Cases')
plt.legend()
plt.show()
```

```python
# How many New cases are arising in various regions?
plt.bar(df['WHO Region'],df['New cases'],label='New Cases')
plt.bar(df['WHO Region'],df['New recovered'],label='New Recovered')
plt.bar(df['WHO Region'],df['New deaths'],label='New Deaths')
plt.title('New Cases')
plt.xlabel('Region')
plt.ylabel('New cases')
plt.legend()
plt.show()
```
